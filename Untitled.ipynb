```python
#Vamos criar uma função que retorna um DataFrame com os segmentos de produtos mais comprados por Estado
def prod_uf(uf):
    prod_uf = olist[olist['customer_state'] == uf]
    prod_uf = prod_uf.groupby(['product_category_name']).price.mean()
    return prod_uf.nlargest(10).plot.bar()

#Município com o maior número de clientes
prod_uf_rj = prod_uf('RJ')
prod_uf_rj
```

```python
rural = geopandas.read_file('RR_Regioes_Rurais2015.shp', encoding='utf-8')
```

```python
rural.head(1)
```

```python
# For para descobrir quantos clientes por municipio e a criação de uma nova coluna com o
# número de clientes.

for index, nome_regiao in rural.iterrows():
  qtd_cliente = len(data[data.intersects(nome_regiao.geometry)])
  rural.loc[index, 'qtd_cliente_rural'] = qtd_cliente
```

```python
data_rj = data[data['SIGLA_UF'] == 'RJ']
```

```python
import folium
```

```python
# Cria o Mapa
fmap = folium.Map(
                location=[-16.1237611, -59.9219642],   
                zoom_start=4,
                control_scale=True,
                prefer_canvas=True)
```

```python
geracoes = folium.map.FeatureGroup()
latitudes = list(data_rj.latitude)
longitudes = list(data_rj.longitude)
labels = list(data_rj.geracoes)
for lat, lng, label in zip(latitudes, longitudes, labels):
    if label == '1960 a 1979 - Geração X':
        folium.Marker(
        location = [lat, lng], 
          popup = label,
          icon = folium.Icon(color='red', icon='o')
         ).add_to(fmap)
    elif label == '1900 a 1940 - Veteranos':
        folium.Marker(
        location = [lat, lng], 
          popup = label,
          icon = folium.Icon(color='black', icon='o')
         ).add_to(fmap)
    elif label == '1940 a 1959 - Baby Boomers ':
        folium.Marker(
        location = [lat, lng], 
          popup = label,
          icon = folium.Icon(color='blue', icon='o')
         ).add_to(fmap)
    elif label == '1980 a 1994 - Geração Y':
        folium.Marker(
        location = [lat, lng], 
          popup = label,
          icon = folium.Icon(color='pink', icon='o')
         ).add_to(fmap)
    else:
        folium.Marker(
          location = [lat, lng], 
          popup = label,
          icon = folium.Icon(color='green', icon='o')
         ).add_to(fmap)
```

```python
fmap.add_child(geracoes)
fmap
```

```python
fmap.save(outfile= "geracoes_rj.html")
```
